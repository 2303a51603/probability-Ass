```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Question 1: Normal Distribution with Mean 75, Std Dev 10
mean_1 = 75
std_dev_1 = 10
data_1 = np.random.normal(mean_1, std_dev_1, 1000)

# Plotting Histogram and PDF
x_1 = np.linspace(mean_1 - 4*std_dev_1, mean_1 + 4*std_dev_1, 1000)
pdf_1 = norm.pdf(x_1, mean_1, std_dev_1)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(data_1, bins=30, density=True, alpha=0.6, color='c', edgecolor='black', label='Histogram')
plt.plot(x_1, pdf_1, 'r-', label='PDF (Normal Curve)')
plt.title('Normal Distribution (Mean=75, Std=10)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
```

```python
# Question 2: BMI Distribution with Mean 20, Std Dev 4
mean_2 = 20
std_dev_2 = 4
x_2 = np.linspace(mean_2 - 4*std_dev_2, mean_2 + 4*std_dev_2, 1000)
pdf_2 = norm.pdf(x_2, mean_2, std_dev_2)

# Probabilities
p_ge_23 = 1 - norm.cdf(23, mean_2, std_dev_2)
p_le_23 = norm.cdf(23, mean_2, std_dev_2)
p_between_23_35 = norm.cdf(35, mean_2, std_dev_2) - norm.cdf(23, mean_2, std_dev_2)

# Plotting BMI PDF with Probability Areas
plt.subplot(1, 2, 2)
plt.plot(x_2, pdf_2, 'g-', label='BMI PDF (Mean=20, Std=4)')
plt.fill_between(x_2, 0, pdf_2, where=(x_2 >= 23), color='r', alpha=0.3, label='BMI ≥ 23')
plt.fill_between(x_2, 0, pdf_2, where=(x_2 <= 23), color='b', alpha=0.2, label='BMI ≤ 23')
plt.fill_between(x_2, 0, pdf_2, where=((x_2 >= 23) & (x_2 <= 35)), color='y', alpha=0.3, label='23 ≤ BMI ≤ 35')
plt.title('BMI Distribution (Mean=20, Std=4)')
plt.xlabel('BMI')
plt.ylabel('Density')
plt.legend()
plt.grid(True)

plt.show()

print(f"P(BMI ≥ 23): {p_ge_23:.4f}")
print(f"P(BMI ≤ 23): {p_le_23:.4f}")
print(f"P(23 ≤ BMI ≤ 35): {p_between_23_35:.4f}")
```
